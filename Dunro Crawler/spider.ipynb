```python
def getIds(page):
    ids = []
    for data in json.loads(page)['data']:
        ids.append(data['uuid'])
    return ids
        
def downloadAllPages(data, pageLimit=9):
    pageIds = []
    for page in range(pageLimit):
        print("fetch -> page", page+1)
        data['page'] = str(page+1)
        r = requests.post('http://dunro.com/api/v1.4/page/search', data=data)
        pageIds += getIds(r.text)
    return pageIds
```
